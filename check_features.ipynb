```python
import pickle

# Load your trained model
model = pickle.load(open("LR.pkl", "rb"))

# Check if model contains stored feature names
print("feature_names_in_ exists?:", hasattr(model, "feature_names_in_"))
print("feature_names_in_ ->", getattr(model, "feature_names_in_", None))

# Check the number of features or coefficients as backup info
print("n_features_in_ ->", getattr(model, "n_features_in_", None))

if hasattr(model, "coef_"):
    import numpy as np
    coef = model.coef_
    print("coef_.shape ->", np.shape(coef))
```

```python
import pickle
features = list(model.feature_names_in_)
with open("features.pkl", "wb") as f:
    pickle.dump(features, f)
print("✅ features.pkl saved successfully with", len(features), "columns.")
```
